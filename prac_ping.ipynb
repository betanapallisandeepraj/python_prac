```python
#for loop with 4 variables to ping all IPs in a list
upper_bound = 255
for i in range(0,upper_bound):
    for j in range(0,upper_bound):
        for k in range(0,upper_bound):
            for l in range(0,upper_bound):
                ip = f"{i}.{j}.{k}.{l}"
                #print(f"pinging {ip}")
                output=!ping -c 1 -t 1 $ip | tail -n 5 | head -n 1
                print(output)
                #!sleep 1
```

```python
#for loop with 4 variables to ping all IPs in a list
upper_bound = 255
for i in range(0,upper_bound):
    for j in range(0,upper_bound):
        for k in range(0,upper_bound):
            for l in range(0,upper_bound):
                ip = f"{i}.{j}.{k}.{l}"
                #print(f"pinging {ip}")
                output=!ping -c 1 -W 100 -i 0.1 $ip | tail -n 5 | head -n 1
                print(output)
                #!sleep 1
```
